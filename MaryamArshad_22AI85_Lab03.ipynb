#Task 3: Scaling & Rotation

image.png

```python


image = cv2.imread("boat-image.jpeg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # convert to RGB for matplotlib
(h, w) = image.shape[:2]

# --- Scaling (200% enlarge) ---
scaled = cv2.resize(image, None, fx=2.0, fy=2.0)

# --- Rotation (90° about center) ---
M = cv2.getRotationMatrix2D((w/2, h/2), 90, 1)  # angle=90°, scale=1
rotated = cv2.warpAffine(image, M, (w, h))


plt.figure(figsize=(15,5))
plt.subplot(1,3,2), plt.imshow(scaled), plt.title("Scaled 200%"), plt.axis("off")
plt.subplot(1,3,3), plt.imshow(rotated), plt.title("Rotated 90°"), plt.axis("off")
plt.show()
```

```python
print("Original shape:", image.shape)
print("Scaled shape:", scaled.shape)
```

#Method

#1.Scaling:
The image was resized using a scaling factor of fx = 2.0 and fy = 2.0, effectively doubling its dimensions.

#2.Rotation:
The scaled image was then rotated by 90° about its center using the rotation matrix from OpenCV.

#Result

1.The scaled image appears enlarged while maintaining the same proportions and aspect ratio.

2.The rotated image is turned by 90° in the counterclockwise direction, with no loss of information except at the borders due to rotation fitting.

#Task 4: Affine Transformation

image.png

```python


# Load triangle image
image = cv2.imread("triangle.jpeg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
(h, w) = image.shape[:2]


pts1 = np.float32([[50, 50], [200, 50], [50, 200]])

pts2 = np.float32([[80, 70], [220, 50], [100, 250]])

# --- Get Affine Transform matrix ---
M_affine = cv2.getAffineTransform(pts1, pts2)

# --- Apply Affine Transform ---
affine = cv2.warpAffine(image, M_affine, (w, h))

plt.figure(figsize=(12,5))
plt.subplot(1,2,1), plt.imshow(image), plt.title("Original Triangle"), plt.axis("off")
plt.subplot(1,2,2), plt.imshow(affine), plt.title("Affine Transformed Triangle"), plt.axis("off")
plt.show()
```

**Rotation:**  rotate the image around a point.

**Scaling:** enlarge or shrink the image.

**Shearing:** slant the image (turn a square into a parallelogram).
