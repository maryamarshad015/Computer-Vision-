```python
import cv2
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread("brighter.jpeg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Split into channels
b, g, r = cv2.split(img_rgb)

# Display channels
plt.subplot(1,3,1)
plt.imshow(r, cmap="Reds")
plt.title("Red Channel")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(g, cmap="Greens")
plt.title("Green Channel")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(b, cmap="Blues")
plt.title("Blue Channel")
plt.axis("off")

plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# Open image
img = Image.open("brighter.jpeg").convert("L")   # Convert to grayscale

# Convert to binary (threshold = 128)
binary = img.point(lambda x: 0 if x < 128 else 255, '1')

# Show results
plt.subplot(1,2,1)
plt.imshow(img, cmap="gray")
plt.title("Grayscale")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(binary, cmap="gray")
plt.title("Binary (PIL)")
plt.axis("off")

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread("brighter.jpeg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Downsample to 100x100
downsampled = cv2.resize(img_rgb, (100, 100), interpolation=cv2.INTER_AREA)

# Upsample back to 400x400
upsampled = cv2.resize(img_rgb, (400, 400), interpolation=cv2.INTER_NEAREST)

# Display results
plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(downsampled)
plt.title("Downsampled (100x100)")
plt.axis("off")


plt.subplot(1,3,3)
plt.imshow(upsampled)
plt.title("Upsampled (400x400)")
plt.axis("off")

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread("brighter.jpeg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply thresholding with different values
threshold_values = [50, 100, 150, 200]
thresh_images = []

for t in threshold_values:
    _, th = cv2.threshold(gray, t, 255, cv2.THRESH_BINARY)
    thresh_images.append((t, th))

# Display results in a clean grid
plt.figure(figsize=(15, 8))

# Original grayscale
plt.subplot(2, 3, 1)
plt.imshow(gray, cmap="gray")
plt.title("Grayscale")
plt.axis("off")

# Thresholded images
for i, (t, th) in enumerate(thresh_images, start=2):
    plt.subplot(2, 3, i)  # arrange in 2x3 grid
    plt.imshow(th, cmap="gray")
    plt.title(f"Threshold {t}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```
